```python
import pandas as pd 
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
titanic_data = pd.read_csv('titanic_train.csv')
disaster_data = titanic_data.drop(columns=['Ticket', 'Cabin', 'Embarked', 'Age', 'Fare'])

disaster_data
```

```python
test_data = pd.read_csv('titanic_test.csv').drop(columns=['Ticket', 'Cabin', 'Embarked', 'Age', 'Fare'])

test_data
```

```python
disaster_data['Sex'] = (disaster_data['Sex'] == 'male').astype(int)
test_data['Sex'] = (test_data['Sex'] == 'male').astype(int)
```

```python
required_features = ['Sex', 'Parch', 'SibSp', 'Pclass']

# x = disaster_data.drop(columns=['Survived', 'Name'])
x = pd.get_dummies(disaster_data[required_features])
y = disaster_data['Survived']

x_test = pd.get_dummies(test_data[required_features])
# x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)

```

```python
# model = xgb.XGBClassifier(
#                          n_estimators=100,
#                          max_depth=10,
#                          learning_rate=0.1,
#                          random_state=0,
#                       )

rf_model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
k_model = KNeighborsClassifier(n_neighbors=2)
# model.fit(x_train, y_train)
rf_model.fit(x, y)


survival_prediction = rf_model.predict(x_test)
```

```python
# Create submission file
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': survival_prediction})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!") 
```
