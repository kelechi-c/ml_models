```python
import pandas as pds 
import xgboost as xgb

from sklearn.model_selection import train_test_split 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
data = pds.read_csv('Salary_Data.csv').dropna()

data
```

```python
data.describe()
```

```python
data['Gender'] = (data['Gender'] == 'Male').astype(int)
```

```python
x = data.drop(columns=['Salary', 'Education Level', 'Job Title'])
y = data['Salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
salary_model = RandomForestRegressor()

salary_model.fit(x_train, y_train)
salary_predictions = salary_model.predict(x_test)

accuracy = r2_score(y_test, salary_predictions)
accuracy
```
