```python
import pandas as pd 
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import classification_report
```

```python
titanic_data = pd.read_csv('titanic_train.csv')
# test_data = pd.read_csv('titanic_test.csv')
disaster_data = titanic_data.drop(columns=['Ticket', 'Cabin', 'PassengerId']).dropna()
disaster_data
```

```python
encoder = OneHotEncoder(handle_unknown='ignore')

data = pd.DataFrame({'Embarked': disaster_data['Embarked']})
encoded_data = encoder.fit_transform(data[['Embarked']]).toarray()

disaster_data.update(encoded_data)

disaster_data['Sex'] = (disaster_data['Sex'] == 'male').astype(int)
```

```python
model = RandomForestClassifier()
x = disaster_data.drop(columns=['Survived', 'Name', 'Embarked'])
y = disaster_data['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

model.fit(x_train, y_train)

survival_prediction = model.predict(x_test)
prediction_accuracy = model.score(x_test, y_test)

prediction_accuracy
```

```python
model_metrics = classification_report(y_test, survival_prediction)

print(model_metrics)
```
